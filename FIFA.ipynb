```python
dataframe = dataframe.drop_duplicates()
dataframe.isna().sum()
```

```python
dataframe.columns
```

```python
dataframe = dataframe.drop(columns=['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf',
       'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb',
       'cb', 'rcb', 'rb', 'gk'])
dataframe.columns
```

```python
dataframe.head()
```

```python
dataframe = dataframe.rename(columns={'bp':'best_position', 'team_&_contract' :'team_and_contract', 'w/f' : 'weak_foot', 'sm' : 'skill_moves', 'a/w' : 'attacking_work_rate', 'd/w' : 'defensive_work_rate', 'ir' : 'international_reputation', 'pac' : 'pace', 'sho' : 'shooting', 'pas' : 'passing', 'dri' : 'dribbing', 'def' : 'defending', 'phy' : 'physicality', 'ova' : 'overall_rating', 'fk_accuracy' : 'free_kick_accuracy', 'gk_diving' : 'goalkeeper_diving', 'gk_handling' : 'goalkeeper_handling', 'gk_kicking' : 'goalkeeper_kicking', 'gk_positioning' : 'goalkeeper_positioning', 'gk_reflexes' : 'goalkeeper_reflexes'})
dataframe.head()
```

```python
dataframe = dataframe.drop(columns=['id', 'foot', 'release_clause', 'position', 'nationality', 'club', 'team_and_contract', 'joined', 'loan_date_end', 'contract'], axis=1)
dataframe.columns
```

```python
dataframe.head()
```

```python
dataframe.best_position.unique()
```

```python
dataframe['best_position'] = dataframe['best_position'].replace({'GK' : 'Goalkeeper', 'CB' : 'Center back', 'LB' : 'Left back', 'RB' : 'Right back', 'CM' : 'Centre midfield', 'CDM' : 'Centre defensive midfield', 'CAM' : 'Centre attacking midfield', 'LM' : 'Left midfield', 'RM' : 'Right midfield', 'ST' : 'Striker', 'CF' : 'Center forward', 'LW' : 'Left winger', 'RW' : 'Right winger', 'LWB' : 'Left Wing Back', 'RWB' : 'Right Wing Back'})
dataframe.best_position.unique()
```

```python
dataframe['name'] = dataframe['name'].str.replace('.', '')
```

```python
dataframe['height'] = dataframe['height'].str.replace('"', '')
dataframe['height'] = dataframe['height'].str.replace("'", '.')
```

```python
dataframe.head()
```

```python
dataframe['weight'] = dataframe['weight'].str.replace('lbs', '')
dataframe.head()
```

```python
dataframe['value'] = dataframe['value'].str.replace('.', '')
dataframe['value'] = dataframe['value'].str.replace('K', '000')
dataframe['value'] = dataframe['value'].str.replace("M", '000000')
dataframe['value'] = dataframe['value'].str.replace("€", '')
dataframe.head()
```

```python
dataframe['wage'] = dataframe['wage'].str.replace('.', '')
dataframe['wage'] = dataframe['wage'].str.replace('K', '000')
dataframe['wage'] = dataframe['wage'].str.replace("M", '000000')
dataframe['wage'] = dataframe['wage'].str.replace("€", '')
dataframe.head()
```

```python
dataframe.weak_foot.value_counts()
```

```python
dataframe['weak_foot'] = dataframe['weak_foot'].str.replace('★', '')
dataframe['skill_moves'] = dataframe['skill_moves'].str.replace('★', '')
dataframe['international_reputation'] = dataframe['international_reputation'].str.replace('★', '')
dataframe.head()
```

```python
dataframe.dtypes
```

```python

dataframe['height']=pd.to_numeric(dataframe['height'], errors='coerce')
dataframe.head(10)
```

```python
dataframe.info()
```

```python
def convert_columns_with_numbers_to_numeric(df):
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            df[column] = pd.to_numeric(df[column], errors='coerce')
    return df
convert_columns_with_numbers_to_numeric(dataframe)
```
